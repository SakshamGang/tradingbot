```python
def signal_generator(df):
    Open = df.open.iloc[-1]
    Close = df.close.iloc[-1]
    previous_Open = df.open.iloc[-2]
    previous_Close = df.close.iloc[-2]
    
    # red candle
    if (Open>Close):
        return 1

    # green candle
    elif (Open<Close):
        return 2
    
    # No clear pattern
    else:
        return 0
```

```python

call=input(str("Enter the symbol for call :"))
put=input(str("Enter the symbol for put :"))
calltight=int(input("Enter qty for tight lots for call option:  "))
callspare=int(input("Enter qty for spare lots for call option:  "))
puttight=int(input("Enter qty for tight lots for put option:  "))
putspare=int(input("Enter qty for spare lots for put option:  "))
```

```python
def get_datacall():
    data = {
    "symbol":call,
    "resolution":"5",
    "date_format":"1",
    "range_from":"2023-11-02",
    "range_to":"2023-11-08",
    "cont_flag":"1"}

    hdata=fyers.history(data=data)
    df = pd.DataFrame.from_dict(hdata['candles'])
    cols = ['datetime','open','high','low','close','volume']
    df.columns=cols
    df['datetime']= pd.to_datetime(df['datetime'],unit="s")
    df['datetime'] = df['datetime'].dt.tz_localize('utc').dt.tz_convert('Asia/Kolkata')
    df['datetime'] = df['datetime'].dt.tz_localize(None)
    df['date'] = df['datetime'].dt.date
    df['time'] = df['datetime'].dt.time
    
    signal = []
    signal.append(0)
    for i in range(1,len(df)):
        dataF = df[i-1:i+1]
        signal.append(signal_generator(dataF))
    #signal_generator(data)
    df["signal"] = signal
    
    goodpair = pd.DataFrame(columns=['Date', 'Time', 'HC', 'LC'])
    #print(goodpairCE)
    
    for i in range(1,len(df)):
        h=0
        l=0
        t=0
        signalfirst = df.signal.iloc[i]
        signalsecond = df.signal.iloc[i-1]
        presenthigh = df.high.iloc[i]
        previoushigh = df.high.iloc[i-1]
        presentlow = df.low.iloc[i]
        previouslow = df.low.iloc[i-1]
        previoustime = df.time.iloc[i-1]
        presenttime = df.time.iloc[i]
        previousdate = df.date.iloc[i-1]
        presentdate = df.date.iloc[i]
    
        #goodpair
        if(signalfirst!=signalsecond):
            if(previoushigh>presenthigh and presentlow<previouslow):
                h = previoushigh
                l = presentlow
                t=1
            elif(previoushigh<presenthigh and presentlow>previouslow):
                h = presenthigh
                l = previouslow
                t=1
        if(abs(h-l)>=0 and abs(h-l)<=30 and t==1):
            goodpair.loc[len(goodpair)] = [previousdate, previoustime, h, l]
    global goodpairCE
    goodpairCE=goodpair
    #return goodpairCE


```

```python
import time
from datetime import date
def onmessage(message):
    t = time.localtime()
    cmin = time.strftime("%M", t)
    csec = time.strftime("%S", t)
    if (int(cmin) % 5 == 0 and int(csec) <9):
        get_datacall()
        get_dataput()
        if (len(goodpairCE)>=1 and goodpairCE.Date.iloc[-1]==date.today()):
            global hc
            hc = goodpairCE.HC.iloc[-1]
            global lc
            lc = goodpairCE.LC.iloc[-1]
        if (len(goodpairPE)>=1 and goodpairPE.Date.iloc[-1]==date.today()):
            global hp
            hp = goodpairPE.HP.iloc[-1]
            global lp
            lp = goodpairPE.LP.iloc[-1]
    if (len(goodpairCE)>=1 and goodpairCE.Date.iloc[-1]==date.today()):
        data_ct = {
            "symbol":call,
            "qty":calltight,
            "type":1,
            "side":1,
            "productType":"BO",
            "limitPrice":0,
            "stopPrice":0,
            "validity":"DAY",
            "disclosedQty":0,
            "offlineOrder":False,
            "stopLoss":lc,
            "takeProfit":hc+15
        }
        response_ct = fyers.place_order(data=data_ct)
        print(f"entry tight slot for call {call}")
        #print(response_ct)
    if (len(goodpairPE)>=1 and goodpairPE.Date.iloc[-1]==date.today()):
        data_pt = {
            "symbol":put,
            "qty":puttight,
            "type":1,
            "side":1,
            "productType":"BO",
            "limitPrice":0,
            "stopPrice":0,
            "validity":"DAY",
            "disclosedQty":0,
            "offlineOrder":False,
            "stopLoss":lp,
            "takeProfit":hp+15
        }
        response_pt = fyers.place_order(data=data_pt)
        print(f"entry tight slot for put {put}")
        #print(response_pt)
    if (len(goodpairCE)>=1 and goodpairCE.Date.iloc[-1]==date.today()):
        data_cs = {
            "symbol":call,
            "qty":callspare,
            "type":1,
            "side":1,
            "productType":"CO",
            "limitPrice":0,
            "stopPrice":0,
            "validity":"DAY",
            "disclosedQty":0,
            "offlineOrder":False,
            "stopLoss":lc,
            "takeProfit":0
        }
        response_cs = fyers.place_order(data=data_cs)
        print(f"entry spare slot for call {call}")
        #print(response_cs)
        target=hc+15
        if(message['symbol']==call and message['ltp']>=target):
            target=target+15
            sl=target-30
            orderId = response_cs['id']
            data_mcs = {
                "id":orderId, 
                "type":1, 
                "limitPrice": sl,
                "qty":callspare
            }
            response_mcs = fyers.modify_order(data=data_mcs)
            print(f"spare slot trailed for call {call} at sl={sl}")
            
    if (len(goodpairPE)>=1 and goodpairPE.Date.iloc[-1]==date.today()):
        data_ps = {
            "symbol":put,
            "qty":putspare,
            "type":1,
            "side":1,
            "productType":"CO",
            "limitPrice":0,
            "stopPrice":0,
            "validity":"DAY",
            "disclosedQty":0,
            "offlineOrder":False,
            "stopLoss":lp,
            "takeProfit":0
        }
        response_ps = fyers.place_order(data=data_ps)
        print(f"entry spare slot for put {put}")
        #print(response_ps)
        target=hp+15
        if(message['symbol']==put and message['ltp']>=target):
            target=target+15
            sl=target-30
            orderId = response_ps['id']
            data_mps = {
                "id":orderId, 
                "type":1, 
                "limitPrice": sl,
                "qty":putspare
            }
            response_mps = fyers.modify_order(data=data_mps)
            print(f"spare slot trailed for put {put} at sl={sl}")

def onerror(message):

    print("Error:", message)


def onclose(message):
  
    print("Connection closed:", message)


def onopen():
    data_type = "SymbolUpdate"
    symbols = [call,put]
    #print(goodpairCE.HC.iloc[-1])
    ofyers.subscribe(symbols=symbols, data_type=data_type)
    ofyers.keep_running()
access_token = client_id + ":" + tk

ofyers = data_ws.FyersDataSocket(
    access_token=access_token,
    log_path="",                     
    litemode=False,
    write_to_file=False,              
    reconnect=True,                 
    on_connect=onopen,               
    on_close=onclose,                
    on_error=onerror,               
    on_message=onmessage             
)
ofyers.connect()
```
